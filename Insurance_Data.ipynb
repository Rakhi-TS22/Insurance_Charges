```python
for i in data.columns:
  if data[i].dtype =='object':
    data[i]= data[i].fillna(data[i].mode()[0])

  else:
    data[i]= data[i].fillna(data[i].mean())
```

```python
le = LabelEncoder()
```

```python
for i in data.columns:
  if data[i].dtype =='object':
    data[i] = le.fit_transform(data[i])
```

```python

# Summary Statistics
print(data.describe())

# Distribution Analysis
numerical_variables = ['age', 'bmi', 'children', 'Claim_Amount', 'past_consultations', 'num_of_steps', 'Hospital_expenditure', 'NUmber_of_past_hospitalizations', 'Anual_Salary', 'charges']
for var in numerical_variables:
    plt.figure(figsize=(8, 5))
    sns.histplot(data[var], kde=True)
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()

# Categorical Variables
categorical_variables = ['sex', 'smoker', 'region']
for var in categorical_variables:
    plt.figure(figsize=(8, 5))
    sns.countplot(data[var])
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.show()

# Correlation Analysis
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Relationships with Target Variable
for var in numerical_variables[:-1]:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=var, y='charges', data=data)
    plt.title(f'{var} vs. Charges')
    plt.xlabel(var)
    plt.ylabel('Charges')
    plt.show()

# Box plots for categorical variables vs. charges
for var in categorical_variables:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=var, y='charges', data=data)
    plt.title(f'{var} vs. Charges')
    plt.xlabel(var)
    plt.ylabel('Charges')
    plt.show()
```

```python
# Multicolinearity VIF value less than 6 than considerin columns for prediction
# Variance Inflation Factor

from statsmodels.stats.outliers_influence import variance_inflation_factor

col_list = []
for i in data.columns:
 if(( data[i].dtypes !='object') &( i !='charges')):
  col_list.append(i)

x= data[col_list]

x
```

```python
vif_data = pd.DataFrame()

vif_data['features'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
```

```python
X = data.drop(['charges'], axis=1)
X
```

```python
y= data['charges']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,accuracy_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score

# Assuming you have already trained your linear regression model and made predictions on the training data
y_train_pred = lr.predict(X_train)

# Calculate R-squared (R2) score on the training data
r2_train = r2_score(y_train, y_train_pred)
print("Training R-squared (R2) score:", r2_train)
```

```python
r2_score(y_pred,y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error (RMSE):", rmse)
```
