```python
# Multicolinearity VIF value less than 6 than considerin columns for prediction
# Variance Inflation Factor

from statsmodels.stats.outliers_influence import variance_inflation_factor

col_list = []
for i in data.columns:
 if(( data[i].dtypes !='object') &( i !='charges')):
  col_list.append(i)

x= data[col_list]

x
```

```python
vif_data = pd.DataFrame()

vif_data['features'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
```

```python
X = data.drop(['num_of_steps', 'charges'], axis=1)
X
```

```python
y= data['charges']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,accuracy_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score

# Assuming you have already trained your linear regression model and made predictions on the training data
y_train_pred = lr.predict(X_train)

# Calculate R-squared (R2) score on the training data
r2_train = r2_score(y_train, y_train_pred)
print("Training R-squared (R2) score:", r2_train)
```

```python
r2_score(y_pred,y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error (RMSE):", rmse)
```
